```python
# Set seed agar hasil konsisten
import pandas as pd 
import numpy as np

np.random.seed(42)

# Generate tanggal
dates = pd.date_range(start="2024-01-01", end="2024-03-31", freq="D")

# Dummy holiday (misal hari besar nasional)
holiday_dates = [
    "2024-01-01",
    "2024-02-10",
    "2024-03-11"
]

df = pd.DataFrame({"date": dates})
df["is_holiday"] = df["date"].isin(pd.to_datetime(holiday_dates)).astype(int)

# Generate sales
df["sales"] = (
    200
    + df["is_holiday"] * 120   # efek holiday
    + np.random.normal(0, 30, len(df))
)

df.head()
```

```python
df["day_of_week"] = df["date"].dt.weekday
df["month"] = df["date"].dt.month
```

```python
df["lag_1"] = df["sales"].shift(1)
df["lag_7"] = df["sales"].shift(7)
```
